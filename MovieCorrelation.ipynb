```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df=pd.read_csv(r'C:\Users\kalli\Downloads\movies.csv')
```

```python
df.head()
```

```python
#Quick missing values checkup
null_value_count = df.isna().sum().sum()
print(null_value_count)
```

```python
#Removing missing values
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df.dtypes
```

```python
#Converting data types
df['budget'] = df['budget'].astype(int)
df['gross'] = df['gross'].astype(int) 
```

```python
#Matching released and year column
df['newyearcolumn'] = df['released'].astype(str)
```

```python
#Dropping unwanted columns created by mistake
df.drop('newcolumnyear',axis=1)
```

```python
#Descending order for revenue
df.sort_values(['gross'], ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Budget vs Gross Visualization
plt.figure(figsize=(10, 7))

plt.scatter(x=df['budget'],y=df['gross'], color='mediumpurple')

plt.title("Budget vs Gross Profit")  

plt.xlabel("Gross Profit")
plt.ylabel("Budget Spent")

plt.show()
```

```python
#Budget vs Gross Visualization using seaborn
plt.figure(figsize=(10, 3))
sns.regplot(x=df['budget'],y=df['gross'],data=df,scatter_kws={"color":"grey"}, line_kws={"color":"black"})
plt.xlim(0, df['budget'].max())
plt.ylim(0, df['gross'].max())
```

```python
#Checking Correlations
df.corr()
```

```python
#Matrix correlation 
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="inferno")
plt.title("Correletion between Gross vs Budget=0.74, Good", size="11", style="italic", color="grey")
plt.show()
```

```python
df_numerized = df.copy()

for col in df_numerized.columns:
    if df_numerized[col].dtype == 'object':
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
df_numerized
```

```python
correlation_matrix=df_numerized.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="inferno")
plt.show()
```

```python
pd.set_option('display.max_rows', None)
x=df_numerized.corr()
corr_pairs=x.unstack()
corr_pairs
```

```python
cor_groups = corr_pairs.sort_values()

cor_groups
```

```python
high_corr = cor_groups[(cor_groups)>0.5]
high_corr
```
